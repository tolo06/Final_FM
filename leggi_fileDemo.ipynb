#**Demo della funzione leggi_file**

##**Definizione della funzione leggi_file**

```python
def leggi_file(nome_file="prova", tipo=".txt", percorso=""):
    try:
        file = open(percorso + nome_file + tipo, 'r')
        contenuto = file.read()
        print(f"\n\n{nome_file}\n\n{contenuto}\n\n")
        file.close()
    except Exception as err:
        print(f"Errore durante la lettura del file: {err}") 
        return False
    else:
        return True
```

##**Chiamata alla funzione e risultato**

```python
leggi_file("prova", ".txt")
```
