##**Demo della funzione apri_immagine**

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

**Definizione della funzione**

```python
def apri_immagine(nome_file="prova", percorso=""):
    try:
        tipo = nome_file[-4:]
        if tipo == ".png":
            img = np.asarray(Image.open(percorso + nome_file))
            plt.imshow(img)
            plt.axis('off')
            plt.show()
        else:
            print("Tipo di file inalido (non '.png')\n")
            return False
    except Exception as err:
        print(f"Errore durante l'apertura dell'immagine: {err}")
        return False
    else:
        return True

   
```

*Chiamata alla funzione e risultato*

```python
apri_immagine("bassotto.png")
```
